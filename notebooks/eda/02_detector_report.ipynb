
# Detector Comparison Report

This notebook aggregates detector metrics, visualizes macro F1 scores, and embeds the latest autoencoder interpretability figures.


```

from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Image

summary_path = Path('reports/detector_compare/detector_comparison.csv')
summary = pd.read_csv(summary_path)
display(summary)
```

```

plt.figure(figsize=(8,4))
sns.barplot(data=summary, x='detector', y='macro_f1', palette='viridis')
plt.title('Macro F1 by Detector')
plt.ylabel('Macro F1')
plt.xlabel('Detector')
plt.ylim(0,1)
plt.tight_layout()
plt.show()
```
