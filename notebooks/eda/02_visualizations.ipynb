
## CNN Detector Probability Timeline


```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cnn_preds = pd.read_csv('models/cnn_classifier/predictions.csv', parse_dates=['timestamp'])
subset = cnn_preds.iloc[:200]
plt.figure(figsize=(10,4))
for label in ['prob_cyber','prob_dropout','prob_normal','prob_spike']:
    sns.lineplot(data=subset, x='timestamp', y=label, label=label)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
